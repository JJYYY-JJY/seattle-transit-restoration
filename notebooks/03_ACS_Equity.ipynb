# Phase 3 ACS Equity Layer

This notebook validates the processed ACS vulnerability layer and visualizes vulnerable tracts for Seattle/King County analysis.

```python
from pathlib import Path
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

ROOT = Path.cwd()
if not (ROOT / 'data').exists() and (ROOT.parent / 'data').exists():
    ROOT = ROOT.parent

geojson_path = ROOT / 'data' / 'processed' / 'acs_demographics_centroids.geojson'
acs = gpd.read_file(geojson_path)

print('Rows:', len(acs))
print('CRS:', acs.crs)
print('Vulnerable tracts:', int(acs['vulnerable'].fillna(False).sum()))
acs.head()
```

```python
summary = acs[['pct_low_income', 'pct_zero_vehicle_households', 'pct_elderly_disabled', 'vulnerability_score']].describe(percentiles=[0.25, 0.5, 0.75])
summary
```

```python
fig, ax = plt.subplots(1, 1, figsize=(11, 11))
acs.plot(
    column='vulnerability_score',
    cmap='YlOrRd',
    linewidth=0.2,
    edgecolor='white',
    legend=True,
    legend_kwds={'label': 'Vulnerability Score'},
    ax=ax,
)
acs[acs['vulnerable'].fillna(False)].boundary.plot(color='black', linewidth=1.0, ax=ax)
ax.set_title('King County Tract Vulnerability Score (ACS)')
ax.set_axis_off()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(11, 11))
acs.boundary.plot(color='lightgray', linewidth=0.2, ax=ax)
vuln = acs[acs['vulnerable'].fillna(False)].copy()
ax.scatter(vuln['centroid_lon'], vuln['centroid_lat'], s=8, c='crimson', alpha=0.7, label='Vulnerable tract centroids')
ax.set_title('Centroids for Vulnerable Tracts (OD Candidate Points)')
ax.set_axis_off()
ax.legend(loc='lower left')
plt.tight_layout()
plt.show()
```
