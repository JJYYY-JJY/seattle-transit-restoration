# Phase 1 Feed Inventory

This notebook validates that Phase 1 data acquisition artifacts are present and readable.

```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path.cwd()
if not (ROOT / 'data').exists() and (ROOT.parent / 'data').exists():
    ROOT = ROOT.parent

RAW = ROOT / 'data' / 'raw'
PROCESSED = ROOT / 'data' / 'processed'
print('Workspace root:', ROOT)
```

```python
manifest_path = PROCESSED / 'feeds_manifest.csv'
if manifest_path.exists():
    manifest = pd.read_csv(manifest_path)
    print('feeds_manifest.csv rows:', len(manifest))
    print(manifest.head(10))
else:
    print('Missing feeds_manifest.csv at', manifest_path)
```

```python
gtfs_current = RAW / 'gtfs_current'
gtfs_archive = RAW / 'gtfs_archive'

current_zips = sorted(gtfs_current.glob('*.zip')) if gtfs_current.exists() else []
archive_zips = sorted(gtfs_archive.glob('*.zip')) if gtfs_archive.exists() else []

print('Current GTFS zip count:', len(current_zips))
for p in current_zips[:5]:
    print('  ', p.name)

print('Archive GTFS zip count:', len(archive_zips))
for p in archive_zips[:10]:
    print('  ', p.name)
```

```python
acs_dir = RAW / 'acs'
if acs_dir.exists():
    acs_files = sorted(acs_dir.glob('*'))
    print('ACS artifacts:', len(acs_files))
    for p in acs_files[:15]:
        print('  ', p.name)
else:
    print('Missing ACS directory')

tract_json = acs_dir / 'acs5_2023_state53_county033_tracts.json'
if tract_json.exists():
    with tract_json.open() as f:
        tracts = json.load(f)
    print('Tract payload rows (incl header):', len(tracts))
```

```python
noaa_dir = RAW / 'noaa'
if noaa_dir.exists():
    noaa_files = sorted(noaa_dir.glob('*'))
    print('NOAA artifacts:', len(noaa_files))
    for p in noaa_files[:15]:
        print('  ', p.name)
else:
    print('Missing NOAA directory')

weather_csv = noaa_dir / 'seattle_daily_weather.csv'
if weather_csv.exists():
    weather = pd.read_csv(weather_csv)
    print('seattle_daily_weather.csv rows:', len(weather))
    print(weather.head())
else:
    print('Missing seattle_daily_weather.csv')
```
